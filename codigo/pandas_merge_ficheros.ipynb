## Pandas merge con 3 ficheros

```python
import pandas as pd
```

```python
path = "../practicas/avanzado2/pandas/merge/"
```

```python
dtPedidos = pd.read_csv(path+"Pedidos.txt", sep=";")
dtPedidos.head()
```

```python
dtEmpresas = pd.read_csv(path+"Empresas.txt", sep=";")
dtEmpresas.head()
```

```python
dtEmpleados = pd.read_csv(path+"Empleados.txt", sep=";")
dtEmpleados.head()
```

```python
dtPedidos_Empresas = pd.merge(dtPedidos, dtEmpresas, left_on='idempresa', right_on='id')
dtPedidos_Empresas.head()
```

```python
dtFinal = pd.merge(dtPedidos_Empresas, dtEmpleados, left_on='idempleado', right_on='id')
dtFinal.head()
```

```python
dtFinal.drop(columns=['idempleado','idempresa','id_x','id_y'], inplace=True)
dtFinal
```

```python
dtFinal.rename(columns={"nombre":"empleado"}, inplace=True)
dtFinal.head()
```

```python
dtFinal = dtFinal[['idpedido','cliente','empleado','cargo','empresa','pais','importe']]
dtFinal.head()
```

```python
dtFinal.to_excel("pedidos.xlsx", index=False, float_format="%.2f", sheet_name="pedidos")
```

```python
g1 = dtFinal['importe'].groupby(dtFinal.pais)
g1
```

```python
print('Media:',g1.mean())
print('Suma:',g1.sum())
print('Mínimo:',g1.min())
print('Máximo:',g1.max())
```

```python
media = dtFinal.groupby(['empleado','empresa']).sum()[['importe']]
media
```
